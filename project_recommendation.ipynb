#### Data Preprocessing and feature engineering

```python
# Converting categorical columns
df['cert_bin'] = df['certification_offered'].map({
    'Yes': 1,
    'No': 0
})
```

```python
df['material_bin'] = df['study_material_available'].map({
    'Yes': 1,
    'No': 0
})
```

```python
diff_map = {
    'Beginner': 0,
    'Intermediate': 1,
    'Advanced': 2
}
df['diff_enc'] = df['difficulty_level'].map(diff_map)
```

```python
from sklearn.preprocessing import MinMaxScaler          # Scales numbers to [0, 1]
```

```python
scaler=MinMaxScaler()
```

```python
# Columns to scale (the raw versions we'll keep for EDA; we create new '_s' scaled versions)
num_cols = ['course_duration_hours', 'course_price', 'enrollment_numbers', 'time_spent_hours']
```

```python
# fit_transform: 'fit' learns the min/max from the data, 'transform' applies the scaling.
# We assign the result to four new columns named with '_s' suffix.
df[['dur_s', 'price_s', 'enroll_s', 'time_s']] = scaler.fit_transform(df[num_cols])
```

```python
print('Scaled columns added. Ranges after scaling:')
print(df[['dur_s','price_s','enroll_s','time_s']].describe().loc[['min','max']].round(4)) #describe gives us the statistical summary of the scaled columns, 
#but we print the 'min' and 'max' of each column to keep it simple. we also round off each to 4 decimal points.
```

#### Feature Engineering

```python
# We create new features that are more meaningful than the raw ones.
# completion_proxy: how much of the course did the user actually consume?= time_spent / course_duration
# This is a proxy for engagement: high value meaning user was engaged, low value meaning dropped out early.
df['completion_proxy'] = (df['time_spent_hours'] / df['course_duration_hours']).clip(0, 1)
#   clip range is from 0 to 1(capped at 1.0 because some users spent > full duration) :anything below 0 becomes 0
#                                                                                      anything above 1 becomes 1
#                                                                                      values between 0 and 1 stay unchanged
```

```python
print('completion_proxy sample:')
print(df['completion_proxy'].describe().round(3))
```

```python
# Build the course feature matrix
# For content-based filtering we need ONE feature vector per course. We will be using hybrid filtering in this project.
# But our dataset has multiple rows per course (one per user who rated it).
# Solution: group by course_id and take the MEAN of every feature.
# This gives a representative profile for each course.

feat_cols = [
    'dur_s',            # Scaled course duration
    'price_s',          # Scaled price
    'enroll_s',         # Scaled enrollment count
    'time_s',           # Scaled avg time spent
    'completion_proxy', # Engagement proxy
    'cert_bin',         # Certification (0/1)
    'material_bin',     # Study material (0/1)
    'diff_enc',         # Difficulty (0/1/2)
    'feedback_score',   # Already in [0,1]
    'previous_courses_taken'  # User experience level
]
```

```python
# groupby('course_id') groups all rows with the same course_id together inside feat_cols
# [feat_cols].mean() computes the column-wise average within each group.
# reset_index() moves course_id back from the index to a regular column.
course_feat = df.groupby('course_id')[feat_cols].mean().reset_index()
```
